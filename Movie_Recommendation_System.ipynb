```python
import ast
def convert1(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l

def convert2(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter == 3:
      break
    l.append(i['name'])
    counter = counter + 1
  return l

def convert3(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == "Director":
      l.append(i['name'])
  return l
```

```python
df['genres']= df['genres'].apply(convert1)
df['keywords']= df['keywords'].apply(convert1)
df['cast'] = df['cast'].apply(convert2)
df['crew'] = df['crew'].apply(convert3)
df.head()
```

**Training the model**

**CountVectorizer**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
len(cv.get_feature_names_out())
```

**Stemming**

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

**Final recommendation after training the model**

```python
 movie_index = new_df[new_df["title"] == "Avatar"].index[0]
```

```python
movie_index
```

```python
def recommend(movie):
  movie_index = new_df[new_df["title"] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True, key=lambda x : x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
